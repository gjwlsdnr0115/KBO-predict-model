# Get Game Stats

Create Dataset with both team's starting lineup

```python
import pandas as pd
import numpy as np
```

### Load Data

```python
hh = pd.read_csv('data_processed/2020/2020_HH.csv', encoding='utf-8')
ht = pd.read_csv('data_processed/2020/2020_HT.csv', encoding='utf-8')
kt = pd.read_csv('data_processed/2020/2020_KT.csv', encoding='utf-8')
lg = pd.read_csv('data_processed/2020/2020_LG.csv', encoding='utf-8')
lt = pd.read_csv('data_processed/2020/2020_LT.csv', encoding='utf-8')
nc = pd.read_csv('data_processed/2020/2020_NC.csv', encoding='utf-8')
ob = pd.read_csv('data_processed/2020/2020_OB.csv', encoding='utf-8')
sk = pd.read_csv('data_processed/2020/2020_SK.csv', encoding='utf-8')
ss = pd.read_csv('data_processed/2020/2020_SS.csv', encoding='utf-8')
wo = pd.read_csv('data_processed/2020/2020_WO.csv', encoding='utf-8')
```

```python
hh.drop('Unnamed: 0', axis=1, inplace=True)
ht.drop('Unnamed: 0', axis=1, inplace=True)
kt.drop('Unnamed: 0', axis=1, inplace=True)
lg.drop('Unnamed: 0', axis=1, inplace=True)
lt.drop('Unnamed: 0', axis=1, inplace=True)
nc.drop('Unnamed: 0', axis=1, inplace=True)
ob.drop('Unnamed: 0', axis=1, inplace=True)
sk.drop('Unnamed: 0', axis=1, inplace=True)
ss.drop('Unnamed: 0', axis=1, inplace=True)
wo.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
teams = {}
teams['HH'] = hh
teams['HT'] = ht
teams['KT'] = kt
teams['LG'] = lg
teams['LT'] = lt
teams['NC'] = nc
teams['OB'] = ob
teams['SK'] = sk
teams['SS'] = ss
teams['WO'] = wo
```

```python
batters = pd.read_csv('dataset_wrangled/2020_팀타자.csv', encoding='utf-8')
pitchers = pd.read_csv('dataset_wrangled/2020_팀투수.csv', encoding='utf-8')
games = pd.read_csv('dataset_wrangled/2020_경기.csv', encoding='utf-8')
```

```python
games
```

```python
game_keys = games['게임키'].tolist()
```

```python
game_keys
```

### Data Processing

```python
# get stat of starting players

def get_game(key, home, away):
    row = []
    home_df = teams[home]
    away_df = teams[away]
    idx_h = home_df[home_df['game_key']==key].index.tolist()[0]
    idx_a = away_df[away_df['game_key']==key].index.tolist()[0]
    for stat in home_df.loc[idx_h]:
        row.append(stat)
    for stat in away_df.loc[idx_a][1:]:
        row.append(stat)
    idx_b = batters[(batters['게임키']==key) & (batters['팀코드']==home)].index.tolist()[0]
    g_b = batters.loc[idx_b]
    row.append(g_b['안타']/g_b['타수'])
    row.append(g_b['득점'])
    idx_p = pitchers[(pitchers['게임키']==key) & (pitchers['팀코드']==home)].index.tolist()[0]
    g_p = pitchers.loc[idx_p]
    row.append(g_p['실점'])
    
    return row
```

```python
total = []
for key in game_keys:
    home = key[10:12]
    away = key[8:10]
    total.append(get_game(key, home, away))
    total.append(get_game(key, away, home))
```

```python
# save result in dataframe
df = pd.DataFrame(total, columns=['game_key','home_p','home_1','home_2','home_3','home_4','home_5','home_6','home_7','home_8','home_9',
                          'away_p','away_1','away_2','away_3','away_4','away_5','away_6','away_7','away_8','away_9',
                          'hit_r','r','l'])
df
```

```python
df.to_csv('data_merged/2020.csv', index=False)
```

### Data finalize

Compile data from all seasons

```python
df_2016 = pd.read_csv('data_merged/2016.csv')
df_2017 = pd.read_csv('data_merged/2017.csv')
df_2018 = pd.read_csv('data_merged/2018.csv')
df_2019 = pd.read_csv('data_merged/2019.csv')
df_2020 = pd.read_csv('data_merged/2020.csv')
```

```python
columns = list(df_2016.columns)
columns.remove('game_key')
columns.remove('hit_r')
columns.remove('r')
columns.remove('l')
```

```python
for col in columns:
    df_2016[col] = df_2016[col].apply(lambda x: eval(x))
for col in columns:
    df_2017[col] = df_2017[col].apply(lambda x: eval(x))
for col in columns:
    df_2018[col] = df_2018[col].apply(lambda x: eval(x))
for col in columns:
    df_2019[col] = df_2019[col].apply(lambda x: eval(x))
for col in columns:
    df_2020[col] = df_2020[col].apply(lambda x: eval(x))
```

```python
# convert each stat into column

all_columns = df_2016.columns.tolist()

total = []
for i in range(len(df_2016)):
    row = []
    for col in all_columns:
        if type(df_2016[col][0])==list:
            for e in df_2016[col][i]:
                row.append(e)
        else:
            row.append(df_2016[col][i])
    total.append(row)
```

```python
# create dataframe

df_cols = ['game_key','hp_1','hp_2','hp_3','hp_4','h1_1','h1_2','h1_3','h1_4','h1_5','h1_6',
           'h2_1','h2_2','h2_3','h2_4','h2_5','h2_6','h3_1','h3_2','h3_3','h3_4','h3_5','h3_6',
           'h4_1','h4_2','h4_3','h4_4','h4_5','h4_6','h5_1','h5_2','h5_3','h5_4','h5_5','h5_6',
           'h6_1','h6_2','h6_3','h6_4','h6_5','h6_6','h7_1','h7_2','h7_3','h7_4','h7_5','h7_6',
           'h8_1','h8_2','h8_3','h8_4','h8_5','h8_6','h9_1','h9_2','h9_3','h9_4','h9_5','h9_6',
           'ap_1','ap_2','ap_3','ap_4','a1_1','a1_2','a1_3','a1_4','a1_5','a1_6',
           'a2_1','a2_2','a2_3','a2_4','a2_5','a2_6','a3_1','a3_2','a3_3','a3_4','a3_5','a3_6',
           'a4_1','a4_2','a4_3','a4_4','a4_5','a4_6','a5_1','a5_2','a5_3','a5_4','a5_5','a5_6',
           'a6_1','a6_2','a6_3','a6_4','a6_5','a6_6','a7_1','a7_2','a7_3','a7_4','a7_5','a7_6',
           'a8_1','a8_2','a8_3','a8_4','a8_5','a8_6','a9_1','a9_2','a9_3','a9_4','a9_5','a9_6',
           'hit_r','r','l']

new_2016 = pd.DataFrame(total, columns=df_cols)
new_2016
```
