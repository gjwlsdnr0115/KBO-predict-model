```python
def get_player_stat(player):
    print(player)
    player_games = batters_by_game[batters_by_game['선수명'] == player]
    player_games.reset_index(inplace=True, drop=True)
    stats = []
    for key in game_keys:
        if key in player_games['게임키'].values:
            idx = player_games[player_games['게임키']==key].index.tolist()[0]
            if idx == 0:
                result = []
                result.append(player_games.loc[idx]['선발'])
                result.append(player_games.loc[idx]['타순'])
                result += [0 for i in range(6)]
                stats.append(result)
            else:
                past = player_games.loc[:idx-1]
                hit_rate = past['안타'].sum() / past['타수'].sum()
                oba = (past['안타'].sum()+past['4구'].sum()+past['고의4구'].sum()+past['사구'].sum())/(past['타수'].sum()+past['4구'].sum()+past['고의4구'].sum()+past['사구'].sum()+past['희비'].sum())
                slg = (past['안타'].sum()+past['2루타'].sum()+2*past['3루타'].sum()+3*past['홈런'].sum())/past['타수'].sum()
                d_hit_rate = past['득점권안타'].sum() / past['득점권타수'].sum()
                d_per_b = past['득점'].sum()/past['타수'].sum()
                t_per_b = past['타점'].sum()/past['타수'].sum()
                result = []
                result.append(player_games.loc[idx]['선발'])
                result.append(player_games.loc[idx]['타순'])
                result += [hit_rate, oba, slg, d_hit_rate, d_per_b, t_per_b]
                for i in range(len(result)):
                    if result[i] != result[i]:
                        result[i] = 0.0
                stats.append(result)
        else:
            if len(stats) == 0:
                stats.append([0, -1, 0, 0, 0, 0, 0, 0])
            else:
                same = stats[-1].copy()
                same[0] = 0
                same[1] = -1
                stats.append(same)
    return stats
```

```python
team_batter_stats = pd.DataFrame()
team_batter_stats['date'] = game_days
team_batter_stats['game_key'] = game_keys


for player in b_list:
    p_stats = get_player_stat(player)
    print(len(p_stats))
    team_batter_stats[player] = p_stats
```

# Pitcher

```python
team = player_2016[player_2016['팀코드'] == 'HH']
team_p, team_b = team[team['포지션'] == '투'], team[team['포지션'] != '투']
```

```python
b_list = team_b['선수명'].tolist()
p_list = team_p['선수명'].tolist()
```

```python
pitchers_by_game = pitcher_2016[pitcher_2016['팀코드'] == 'HH']
pitchers_by_game.reset_index(inplace=True, drop=True)
pitchers_by_game
```

```python
game_keys = pitchers_by_game['게임키'].unique().tolist()
game_keys
```

```python
game_days = []
for key in game_keys:
    idx = pitchers_by_game[pitchers_by_game['게임키']==key].index.tolist()[0]
    date = pitchers_by_game.loc[idx]['일자']
    game_days.append(date)
game_days
```

```python
pitchers_by_game.columns
```

```python
def get_pitcher_stat(player):
    print(player)
    player_games = pitchers_by_game[pitchers_by_game['선수명'] == player]
    player_games.reset_index(inplace=True, drop=True)
    stats = []
    for key in game_keys:
        if key in player_games['게임키'].values:
            idx = player_games[player_games['게임키']==key].index.tolist()[0]
            if idx == 0:
                result = []
                result.append(player_games.loc[idx]['선발'])
                result.append(player_games.loc[idx]['구원'])
                result += [0 for i in range(4)]
                stats.append(result)
            else:
                past = player_games.loc[:idx-1]
                era = past['자책점'].sum() / past['이닝*3'].sum() * 27
                whip = (past['안타'].sum() + past['4구'].sum()) / (past['이닝*3'].sum() / 3)
                if past['4구'].sum() == 0:
                    k_b = past['삼진'].sum()
                else:
                    k_b = past['삼진'].sum() / past['4구'].sum()
                aver_inning = (past['이닝*3'].sum() / 3) / len(past)
                result = []
                result.append(player_games.loc[idx]['선발'])
                result.append(player_games.loc[idx]['구원'])
                result += [era, whip, k_b, aver_inning]
                for i in range(len(result)):
                    if result[i] != result[i]:
                        result[i] = 0.0
                    if np.isinf(result[i]):
                        result[i] = 10
                stats.append(result)
        else:
            if len(stats) == 0:
                stats.append([0, 0, 0, 0, 0, 0])
            else:
                same = stats[-1].copy()
                same[0] = 0
                same[1] = 0
                stats.append(same)
    return stats
```

```python
team_pitcher_stats = pd.DataFrame()
team_pitcher_stats['date'] = game_days
team_pitcher_stats['game_key'] = game_keys


for player in p_list:
    p_stats = get_pitcher_stat(player)
    print(len(p_stats))
    team_pitcher_stats[player] = p_stats
```

```python
team_pitcher_stats.to_csv('data_processed/2016/2016_HH_p.csv')
```

```python
def get_win_rate(game_key):
    idx = team_pitch[team_pitch['게임키']==game_key].index.tolist()[0]
    if idx == 0:
        return 0
    else:
        past = team_pitch.loc[:idx-1]
        counts = past['결과'].value_counts()
        wins = counts['W']
        lose = counts['L']
        rate = wins / (win+lose)
        return rate
```
